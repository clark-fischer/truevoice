```python
import geopandas as gpd
import matplotlib.pyplot as plt
import zipfile
```

```python
shp_path = 'nv_2020_final/nv_2020_final.shp'
cpg_path = 'nv_2020_final/nv_2020_final.cpg'
dbf_path = 'nv_2020_final/nv_2020_final.cdf'
shx_path = 'nv_2020_final/nv_2020_final.shx'
prj_path = 'nv_2020_final/nv_2020_final.prj'
geojson_path = 'nv_2020_final/nv_2020_final.geojson'
```

```python
shapefile = gpd.read_file(shp_path)
geojson = gpd.read_file(geojson_path)



```

```python
fig, ax = plt.subplots(figsize=(10, 10))

shapefile.plot(ax=ax, color='lightblue', edgecolor='black', alpha=0.6, label="Shapefile")
geojson.plot(ax=ax, color='lightgreen', edgecolor='black', alpha=0.7, label="GeoJSON")
plt.legend()
plt.title('Shapefile and GeoJSON Overlay')
plt.show()
```
