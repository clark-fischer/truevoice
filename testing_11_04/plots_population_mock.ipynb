Box and whisker plot MMD

```python
def MMD_box_and_whisker_plot(data, title, x_label, y_label):
    plot_data = [[district["min"], district["q1"], district["median"], district["q3"], district["max"]] for district in data["ensembles"]]
    representatives_per_district = [district["representatives"] for district in data["ensembles"]]
    
    plt.figure(figsize=(8, 6))
    plt.boxplot(plot_data, labels=[f"District {j+1} (Reps: {representatives_per_district[j]})" for j in range(len(plot_data))],
            patch_artist=True, medianprops=dict(color="black"))
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.grid(True)
    plt.tight_layout()
    plt.show()


```

SMD Box and Whisker plot

```python
def SMD_box_and_whisker_plot(data,title, x_label, y_label):
    ensemble_data = [[district["min"], district["q1"], district["median"], district["q3"], district["max"]]
                     for district in data["ensembles"]]
    enacted_values = [district["enacted_value"] for district in data["district_id"]]
    plt.figure(figsize=(10, 6))
    plt.boxplot(ensemble_data, labels=[f"District {j+1}" for j in range(len(ensemble_data))],
                patch_artist=True, medianprops=dict(color="black"))
    for i, value in enumerate(enacted_values):
        plt.scatter(i + 1, value, color="red", zorder=5, label="Enacted Plan" if i == 0 else "")
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
def opportunity_representatives_plot(data, title, x_label, y_label):
    opportunity_representatives = [d["opportunity_representatives"] for d in data["ensembles"]]

    plt.figure(figsize=(8, 5))
    plt.hist(opportunity_representatives, color='#7ff5b8',align='left',rwidth=1)
    seat_share = f"Average Seat Share: {data['average_seat_share'] * 100}%"
    vote_share = f"Vote Share: {data['vote_share'] * 100}%"
    plt.gcf().text(0.65, 0.8, f"{seat_share}\n{vote_share}", fontsize=10,
                   bbox=dict(facecolor='white', alpha=0.8, edgecolor='black'))
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.show()
    
def opportunity_district_plot(data, title, x_label, y_label):
    opportunity_districts = [d["opportunity_districts"] for d in data["ensembles"]]
  
    plt.figure(figsize=(8, 5))
    plt.hist(opportunity_districts, color='#7ff5b8')
    seat_share = f"Average Seat Share: {data['average_seat_share'] * 100}%"
    vote_share = f"Vote Share: {data['vote_share'] * 100}%"
    plt.gcf().text(0.65, 0.8, f"{seat_share}\n{vote_share}", fontsize=10,
                   bbox=dict(facecolor='white', alpha=0.8, edgecolor='black'))
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.show()

def party_split_plot(data, title, x_label, y_label):
    
    dem_values = [d["DEM"] for d in data["ensembles"]]

    
    #mean = 0.5
    #std_dev = 0.1  # Adjust standard deviation to control the spread
    #dem_values = np.random.normal(mean, std_dev, 5000)
    #dem_values = np.clip(dem_values, 0, 1)

    bins = [i / 200 for i in range(201)] 
    plt.figure(figsize=(10, 6))
    plt.hist([val for val in dem_values if val < 0.5], bins=bins, color='#FF0000', alpha=0.7, label="Below 50% (REP)")
    plt.hist([val for val in dem_values if val >= 0.5], bins=bins, color='#0015bc', alpha=0.7, label="Above 50% (DEM)")
    seat_share = f"Average Seat Share: {data['average_seat_share'] * 100}%"
    vote_share = f"Vote Share: {data['vote_share'] * 100}%"
    plt.gcf().text(0.65, 0.8, f"{seat_share}\n{vote_share}", fontsize=10,
                   bbox=dict(facecolor='white', alpha=0.8, edgecolor='black'))
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.show()
def party_split_bar_plot(data, title, x_label, y_label):
    
    diff_values = [d["DEM"] - d["REP"] for d in data["ensembles"]]
    
    bins = [i / 100 for i in range(101)] 
    plt.figure(figsize=(10, 6))
    plt.hist([val for val in diff_values if val < 0],bins=100, color='#FF0000', alpha=0.7, label="Below 50% (REP)")
    plt.hist([val for val in diff_values if val >= 0],bins=100, color='#0015bc', alpha=0.7, label="Above 50% (DEM)")
    
    seat_share = f"Average Seat Share: {data['average_seat_share'] * 100}%"
    vote_share = f"Vote Share: {data['vote_share'] * 100}%"
    plt.gcf().text(0.65, 0.8, f"{seat_share}\n{vote_share}", fontsize=10,
                   bbox=dict(facecolor='white', alpha=0.8, edgecolor='black'))
    plt.xlim(-1, 1)
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.show()



    
```

```python
def vote_share_seat_share_plot(data, title, x_label, y_label):
    vote_shares = [d["vote_share"] for d in data["ensemble"]]
    seat_shares = [d["seat_share"] for d in data["ensemble"]]

    
    vote_shares = np.linspace(0, 1, 100)  
    seat_shares = 1 / (1 + np.exp(-10 * (vote_shares - 0.5)))  


    plt.figure(figsize=(10, 6))
    plt.plot(vote_shares, seat_shares, color='black', linewidth=2)
    plt.scatter(
        [data["vote_share"]], 
        [data["smd_seat_share"]], 
        color='red', s=100, label="SMD Vote/Seat Share"
    )
    plt.scatter(
        [data["vote_share"]], 
        [data["mmd_seat_share"]], 
        color='blue', s=100, label="MMD Vote/Seat Share"
    )

    plt.text(0.05, 0.9, f"Symmetry: {data['symmetry']}", fontsize=9, color='black')
    plt.text(0.05, 0.85, f"Bias: {data['bias']}", fontsize=9, color='black')
    plt.text(0.05, 0.8, f"Responsiveness: {data['responsiveness']}", fontsize=9, color='black')

    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Mock data for the S-Curve parameters
vote_share = 0.55  # Overall vote share for the party
seat_share = 0.50  # Overall seat share for the party
symmetry = 0.45    # Symmetry of the curve
bias = 1.2         # Bias metric
responsiveness = 0.85  # Responsiveness metric

# Generate vote share range for the curve
vote_share_range = np.linspace(0, 1, 100)

# Define the S-Curve using logistic-like function with bias and responsiveness
def s_curve(vote_share_range, center, steepness):
    return 1 / (1 + np.exp(-steepness * (vote_share_range - center)))

# Adjusting center based on symmetry and bias
center = 0.5 + (bias / 100)
steepness = 10 * responsiveness  # Adjust steepness based on responsiveness
seat_share_curve = s_curve(vote_share_range, center, steepness)

# Plot the S-Curve
plt.figure(figsize=(8, 5))
plt.plot(vote_share_range, seat_share_curve, label="S-Curve (Vote-Seat Share)", color="blue")

# Highlight the given vote share and seat share point
plt.scatter([vote_share], [seat_share], color="red", label="Given Plan Point", s=100, edgecolor="black")

# Add annotations for symmetry, bias, and responsiveness
plt.annotate(f"Symmetry: {symmetry}", xy=(0.05, 0.85), xycoords="axes fraction")
plt.annotate(f"Bias: {bias}", xy=(0.05, 0.80), xycoords="axes fraction")
plt.annotate(f"Responsiveness: {responsiveness}", xy=(0.05, 0.75), xycoords="axes fraction")

# Add reference lines
plt.axhline(0.5, color="gray", linestyle="--", linewidth=0.8, label="50% Seat Share")
plt.axvline(0.5, color="gray", linestyle="--", linewidth=0.8, label="50% Vote Share")

# Labels, title, and legend
plt.title("S-Curve for Vote-Seat Share with Given Metrics")
plt.xlabel("Vote Share")
plt.ylabel("Seat Share")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()

# Show the plot
plt.show()
```
