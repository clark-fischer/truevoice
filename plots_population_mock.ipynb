```python
def MMD_box_and_whisker_plot(data, title, x_label, y_label):
    plot_data = [[district["min"], district["q1"], district["median"], district["q3"], district["max"]] for district in data["ensembles"]]
    representatives_per_district = [district["representatives"] for district in data["ensembles"]]
    
    plt.figure(figsize=(8, 6))
    plt.boxplot(plot_data, labels=[f"District {j+1} (Reps: {representatives_per_district[j]})" for j in range(len(plot_data))],
            patch_artist=True, medianprops=dict(color="black"))
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.grid(True)
    plt.tight_layout()
    plt.show()


```

SMD Box and Whisker plot

```python
def SMD_box_and_whisker_plot(data,title, x_label, y_label):
    ensemble_data = [[district["min"], district["q1"], district["median"], district["q3"], district["max"]]
                     for district in data["ensembles"]]
    enacted_values = [district["enacted_value"] for district in data["district_id"]]
    plt.figure(figsize=(10, 6))
    plt.boxplot(ensemble_data, labels=[f"District {j+1}" for j in range(len(ensemble_data))],
                patch_artist=True, medianprops=dict(color="black"))
    for i, value in enumerate(enacted_values):
        plt.scatter(i + 1, value, color="red", zorder=5, label="Enacted Plan" if i == 0 else "")
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
def opportunity_representatives_plot(data, title, x_label, y_label):
    opportunity_representatives = [d["opportunity_representatives"] for d in data["ensembles"]]

    plt.figure(figsize=(8, 5))
    plt.hist(opportunity_representatives, color='#7ff5b8',align='left',rwidth=1)
    seat_share = f"Average Seat Share: {data['average_seat_share'] * 100}%"
    vote_share = f"Vote Share: {data['vote_share'] * 100}%"
    plt.gcf().text(0.65, 0.8, f"{seat_share}\n{vote_share}", fontsize=10,
                   bbox=dict(facecolor='white', alpha=0.8, edgecolor='black'))
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.show()
    
def opportunity_district_plot(data, title, x_label, y_label):
    opportunity_districts = [d["opportunity_districts"] for d in data["ensembles"]]
  
    plt.figure(figsize=(8, 5))
    plt.hist(opportunity_districts, color='#7ff5b8')
    seat_share = f"Average Seat Share: {data['average_seat_share'] * 100}%"
    vote_share = f"Vote Share: {data['vote_share'] * 100}%"
    plt.gcf().text(0.65, 0.8, f"{seat_share}\n{vote_share}", fontsize=10,
                   bbox=dict(facecolor='white', alpha=0.8, edgecolor='black'))
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.show()

def party_split_plot(data, title, x_label, y_label):
    
    dem_values = [d["DEM"] for d in data["ensembles"]]

    
    #mean = 0.5
    #std_dev = 0.1  # Adjust standard deviation to control the spread
    #dem_values = np.random.normal(mean, std_dev, 5000)
    #dem_values = np.clip(dem_values, 0, 1)

    bins = [i / 200 for i in range(201)] 
    plt.figure(figsize=(10, 6))
    plt.hist([val for val in dem_values if val < 0.5], bins=bins, color='#FF0000', alpha=0.7, label="Below 50% (REP)")
    plt.hist([val for val in dem_values if val >= 0.5], bins=bins, color='#0015bc', alpha=0.7, label="Above 50% (DEM)")
    seat_share = f"Average Seat Share: {data['average_seat_share'] * 100}%"
    vote_share = f"Vote Share: {data['vote_share'] * 100}%"
    plt.gcf().text(0.65, 0.8, f"{seat_share}\n{vote_share}", fontsize=10,
                   bbox=dict(facecolor='white', alpha=0.8, edgecolor='black'))
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.show()
def party_split_bar_plot(data, title, x_label, y_label):
    
    diff_values = [d["DEM"] - d["REP"] for d in data["ensembles"]]
    print(diff_values)
    bins = [i / 100 for i in range(101)] 
    plt.figure(figsize=(10, 6))
    plt.hist([val for val in diff_values if val < 0],bins=100, color='#FF0000', alpha=0.7, label="Below 50% (REP)")
    plt.hist([val for val in diff_values if val >= 0],bins=100, color='#0015bc', alpha=0.7, label="Above 50% (DEM)")
    
    seat_share = f"Average Seat Share: {data['average_seat_share'] * 100}%"
    vote_share = f"Vote Share: {data['vote_share'] * 100}%"
    plt.gcf().text(0.65, 0.8, f"{seat_share}\n{vote_share}", fontsize=10,
                   bbox=dict(facecolor='white', alpha=0.8, edgecolor='black'))
    plt.xlim(-1, 1)
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.show()



    
```

```python
districts = [
    {"district_id": 1,
     "vote_share": 20, 
     "seat_share": 80},
    {"district_id": 2, 
     "vote_share": 40, 
     "seat_share": 60},
    {"district_id": 3, "vote_share": 45, "seat_share": 55},
    {"district_id": 4, "vote_share": 45, "seat_share": 55},
    {"district_id": 5, "vote_share": 20, "seat_share": 80},
    {"district_id": 6, "vote_share": 40, "seat_share": 60},
    {"district_id": 7, "vote_share": 45, "seat_share": 55},
    {"district_id": 8, "vote_share": 45, "seat_share": 55},
]

vote_share_intervals = np.linspace(0, 1, 11)


def calculate_seat_shares(vote_share_intervals, districts, party_key):
    seat_shares = []
    for vote_share in vote_share_intervals:
        seats_won = 0
        for district in districts:
            total_votes = district["vote_share"] + district["seat_share"]
            party_votes = vote_share * total_votes
            other_votes = (1 - vote_share) * total_votes
            if party_votes > other_votes:
                seats_won += 1
        seat_shares.append((seats_won / len(districts)) * 100)
    return seat_shares
def calculate_symmetry(vote_share_intervals, seat_shares):

    total_intervals = len(vote_share_intervals)
    symmetry_values = []
    
    for i in range(total_intervals // 2 + 1):  # Only need to calculate for half the intervals
        seat_share_at_x = seat_shares[i]
        seat_share_at_1_minus_x = seat_shares[total_intervals - i - 1]
        
        # Symmetry is the absolute difference between seat shares at x and (1-x)
        symmetry_values.append(abs(seat_share_at_x - (100 - seat_share_at_1_minus_x)))
    
    # Average the symmetry values to get an overall symmetry measure
    symmetry = np.mean(symmetry_values)
    return symmetry
# Calculate seat shares for Democrats
dem_seat_shares = calculate_seat_shares(vote_share_intervals, districts, "vote_share")
print(dem_seat_shares)

# Calculate responsiveness, bias, and symmetry
def calculate_metrics(vote_share_intervals, seat_shares):
    responsiveness = (seat_shares[6] - seat_shares[4]) / (vote_share_intervals[6] - vote_share_intervals[4])  # Change near 50%
    bias = seat_shares[5] - 50  # Seat share at 50% vote share
    symmetry = calculate_symmetry(vote_share_intervals,seat_shares)
    return responsiveness, bias, symmetry

responsiveness, bias, symmetry = calculate_metrics(vote_share_intervals, dem_seat_shares)

# Plot the S-curve for Democrats
plt.figure(figsize=(10, 6))
plt.plot(vote_share_intervals * 100, dem_seat_shares, label="Democrats", marker="o")
plt.title("Vote Share vs Seat Share (Democrats) - Colorado")
plt.xlabel("Vote Share (%)")
plt.ylabel("Seat Share (%)")
plt.grid(True)
plt.legend()
plt.show()
```
